# 20.8 DataFrame Groupby kullanımı

```python
#belli ekrarlanana verilere göre gruplama yapabilmek amaçtır
```

```python
import pandas as pd
import numpy as np
```

```python
personeller = {
    "Çalışan": ["ahmet yılmaz","can ertürk","hasan korkmaz","cenk saymaz","ali kaptan","kerem kalkan","cem alan"],
    "Departman": ["insan kaynakları", "bilgi işlem","muhasebe","insan kaynakları","bilgi işlem","muhasebe","insan kaynakları"],
    "Yaş": [30,25,45,50,23,34,42],
    "Semt": ["kadıköy","tuzla","maltepe","tuzla","maltepe","tuzla","kadıköy"],
    "Maaş": [5000,3000,4000,3500,2750,6500,4500]
}

df = pd.DataFrame(personeller)

df
```

```python
result = df["Maaş"].sum() #7 personelin maaşları toplandı
result = df.groupby("Departman").groups #groups özelliği ile hangi elemanar gruplanmış görebilirim, gruplanan verilerin indekslerini yanına yazar çıktıda
result = df.groupby(["Departman","Semt"]).groups #bir liste içinde gönderirim ve iki tanesi aynı olanları gruplayacak yani departman ve semti tutanları gruplayacak

result
```

```python
for name, group in df.groupby("Semt"): #gruplanmış verileri bana semtin altında açıklayarak verdi
    print(name)
    print(group)    
```

```python
for name, group in df.groupby("Departman"): #departmanda gruplanmış verileri bana departman adları altında açıklayarak verdi
    print(name)
    print(group)
```

```python
result = df.groupby("Semt").get_group("kadıköy") #get_group özelliğiyle semtlerin içinden sadece kadıköyde yaşayanları seçip getirebiliyorum
result = df.groupby("Departman").get_group("muhasebe") #sadece muhasebe
result
```

```python
result = df.groupby("Departman").sum() # departmandaki kişilerin yaş ve maaş toplamı geldi çünkü kolon belirtmedim 

result = df.groupby("Departman").mean() #ortalamaları

result = df.groupby("Departman")["Maaş"].mean() #sadece maaş ortalamları hesaplanır. istediğim kolonu sonra yanına yazarsam

result = df.groupby("Semt")["Yaş"].mean() #semte göre oturan kiişlerin yaş ortalamaları

result = df.groupby("Semt")["Maaş"].sum() #semte göre maaş toplamları

result = df.groupby("Semt")["Çalışan"].count() #semte göre o semtten kaç kişi çalışıyor

result = df.groupby("Departman")["Yaş"].max() # departmana göre en çok hangi yaş çalışıyor

result = df.groupby("Departman")["Maaş"].min() #departmana göre en az alınan maaşlar

result = df.groupby("Departman")["Maaş"].max()["muhasebe"] # yine sonuna istediğim bölümü eklersem artık tüm departmanlar değil sadece muhasebe departmanında en yüksek maaşı alan gelecek

result = df.groupby("Departman").agg(np.mean) #agg metodu ile npden ort hesaplatabilirim

result = df.groupby("Departman")["Maaş"].agg([np.sum,np.mean,np.max,np.min]) # agg kullanmamızın sebebi birden fazla parametre verebiliyor olmam burada tüm işlemleri tek seferde yaptırabildim

result = df.groupby("Departman")["Maaş"].agg([np.sum,np.mean,np.max,np.min]).loc["muhasebe"] #bunu sadece tek bir bilgi için yapmak istersem ise loc belirterek yapabilirim , burada işlem sadece muhasebe birimie göre yapılmış oldu


result
```

```python
result = df.drop("column1", axis = 1)  # drop kolon silmek için kullanılır
result = df.drop(["column1","column2"], axis = 1)
result = df.drop("a", axis = 0) #satır silmek istersem direkt satırı yazıp axis 0 yapmam lazım
result = df.drop(["a","b","h"], axis = 0) # birden fazla silmek için yine liste iiçnde atıyorum

result = df.isnull() #isnull ile nan olan değerleri görebilirim True olanlar nan değerlerdir
result = df.notnull() #notnull da tam tersi nanlar false olur
result = df.isnull().sum() #nun değerlerin toplamı
result =df["column1"].isnull().sum() #sadece kolon1dei null değerlere bakmak istersem kolon ismi belirtmem gerekiyor

newcolumn = [np.nan,30,np.nan,51,np.nan,30,np.nan,10] # yeni kolon oluşturuyorum bunun için bilgilerimi yazdım 8 satırlık bir veri
df["column4"] = newcolumn #oluşturduğum değerleri dfde kolon4e atarak yeni kolonu eklemiş oldum


result =df[df["column1"].isnull()]["column1"] #kolon birdeki nan değerleri yazdırmak istersem df içine yazıyorum
result =df[df["column1"].notnull()]["column1"] #kolon1deki nun olmayan değerleri görmek için


result = df.dropna() #nan değer içeren kolon ve satırları sil demek dropna varsayılan axis değeri 0dır satıra göre siler. çalıştırırsam sadece değerlerden oluşan satırlar gelir nan olan satır değerlerini siler
result = df.dropna(axis=1) #sadece sütuna göre o sütunda nan değer varsa siler
result = df.dropna(how = "any") #how parametresi var ve any dersem herhangi bir kolonda nan değer bulursa siler
result = df.dropna(how = "all") #all dersem de sadece tüm kolonda nansa karşıma gelmez
result = df.dropna(subset = ["column1","column2"], how = "all") #subset özelliği ile verdiğim bilgileri şu kolonda ara diyebiliyorum örneğin nan verileri bu kolonda ara
result = df.dropna(subset = ["column1","column2"], how = "any")
result = df.dropna(thresh = 2) #thresh özelliği ile iki tane normal veri varsa bu kayıtları silme diyorum burada yani thresh parametresiyle en az şu kadar normal veri varsa silme diyebilirim
result = df.dropna(thresh = 4)
result = df.dropna(thresh = 3)# en az sayıda normal veri

result = df.fillna(value = "no input") #fillna metodu benim için boş verileri doldurur ve içine parametre olarak bir value değeri veririm yani dolduracağı veriyi
result = df.fillna(value = 1)
#******************** kolonları ortlama ile dolduracağım
result =df.sum() # verilerin toplamı
result =df.sum().sum() # verilerin tüm toplamı
result = df.size #toplam kaç değer var
result = df.isnull().sum() # boş değerlerin kolon toplamı 
result = df.isnull().sum().sum() #boş değerlerin tüm toplamı


def ortalama(df): #ortalma hesaplayacak fonksiyon
    toplam = df.sum().sum()
    adet = df.size - df.isnull().sum().sum()
    return toplam / adet

result = df.fillna(value = ortalama(df)) #ortlamayı gelecek boş veriler içine yazacak


result
#df
```

# 20.10 pandas ile string fonksiyonları kullanımı

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("all_seasons.csv")
data
```

```python
data.dropna(inplace = True) #nan değerleri kayıtlar içinden sildim
```

```python
data.columns #kolon isimlerine baktım
```

```python
data["player_name"] = data["player_name"].str.upper() #oyuncu isimlerindeki tüm harfleri büyütmek istiyorum burada önce str yazıyorum str fonksiyonlarına ulaşmak iiçn sonra upper metodunu kullanıyorum
data.head(10) #player name kolonu büyük harflerle yazıldı

data["player_name"] = data["player_name"].str.lower()
data
```

```python
data["index"] = data["player_name"].str.find("a") #
data #yeni kolon hazırladım en sağ kolonda yeni bir kolon çıkacak ve bana player name kolonunda bulduğu a harfinnin index numarlarını verecek
```

```python
data = data[data["player_name"].str.contains("jordan")]
#data["player_name"].str.contains("jordan")

data.head(23) # içerisinde sadece jordan geçen kayıtları getirir kodu data iiçne yazıp verdim
```

```python
data = data.player_name.str.replace(" ","-")#.str.replace("*"," ") #devam etmek istersem yine .str.replace ile devam edebilirim
data.head(10) #boşluk karakterlerini - ile değiştirdim
```

```python
data[["Firsat name","Last name"]] = data["player_name"].loc[data["player_name"].str.split().str.len()==2].str.split(expand=True)
# yeni kolon oluşturup player namedeki siimleri firsat ve last name kolonlarına ayırdım str yazdım çünkü str metodu kullanmak istiyorum split ile boşluklardan böldüm ve len ile eğer playername kısmı iki ise expland True diyerek ayrı kolonlara gitmesini sağlıyorum
```
