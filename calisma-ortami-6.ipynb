```python
# veri setimi yüklüyorum. titanic

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
!pip install missingno
import missingno as msno
from datetime import date
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import LocalOutlierFactor #çok değişkenli ayıklama yöntemi bunkarı gptye sor
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, StandardScaler, RobustScaler #standartlaştırma ve dönüştürme fonksiyonları


```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

burada kabin değişkeni vardı bazı değerlerde nan bazı değerlerde nan olmama durumu vardı. şimdi öyle bir işlem yapmak istiyorum ki nan olan yerlere 1 nan olmayan yerlere 0 yazmak istiyorum. normalde kabin değişkeninde çok fazla eksik değer olduğu için silebilirdik. ama şunu merak ediyorum bu nan yerlere 0 diğerlerine 1 yazayım ve gözlemleyeyim bakayım mesela bağımlı değişken ile bakayım bakalım gerekten anlamsız mı sileyim mi diye fikir edinemye çalışıp düşünüyorum. işlemlere geçiyorum.

```python
df["NWE_CABIN_BOOL"] = df["Cabin"].notnull().astype("int")
# cabin değişkeni notnull(null değil mi) yani dolu mu diye soruyorum sonuç olarak buradan true false değerleri geliyor olacak ve 
# bu true false değerleri 1 ve 0 ı temsil edeceğinden dolayı astype int diyerek bunları integrea çeviriyorum.
```

```python
df.head() #new_cabin_bool isminde yeni bir değişkene atadığım için bu isimde bir değişken geldi
# nan olan yere 0 dedi dolu olan yere 1 dedi.
#şu anda anlamsız olabilecek bir değişken benim için artık daha anlamlı bir hale geldi.
```

```python
#şimdi de bağımlı değişkenime göre ortalama alacağım
#yeni oluşturdğum değişkene göre groupby alıp survivedın ortalamasını al diyorum
df.groupby("NWE_CABIN_BOOL").agg({"Survived": "mean"})
```

```python
#dikkat kabini dolu olanların hayatta kalma oranı %66 boş olanların ise %29
#yani nan olanlara göre daha yüksek.
#daha önceden benim için önemsiz gibi duran bu değişken artık benim için oldukça önemli bir hale geldi
# çünkü artık içerisinde önemki bir bilgi barındırıyor.
# artık şu yorumu yapabiliyorum artık bağımlı değişkenim survived açısından değerlendirdiğimde cabini dolu olan kişilerin hayatta kalma oranı boş olanlara göre 2 ka tdaha fazla
# bu durumda bunların arasında gerçekten bir ilişki olabilir mi diye düşünüyorum.
#buraya bi istatistiki test yapacağım. oran testi yapacağım.
# iki grubun oranını kıyaslama testi yapacağım.
#yeni oluşturduğum bir feature var bu featureın bağımlı değişken ile aralarındaki ilişkiyi merak ediyorum ve bunun için oran testi yapıyorum
```

```python
from statsmodels.stats.proportion import proportions_ztest #ora ntstini import ettim
```

```python
test_stat, pvalue = proportions_ztest(count = [df.loc[df["NWE_CABIN_BOOL"] == 1, "Survived"].sum(),
                                              df.loc[df["NWE_CABIN_BOOL"] == 0, "Survived"].sum()],
                                     
                                     nobs = [df.loc[df["NWE_CABIN_BOOL"] == 1, "Survived"].shape[0],
                                            df.loc[df["NWE_CABIN_BOOL"] == 0, "Survived"].shape[0]])

print("Test Stat = %.4f, p-value = %.4f" % (test_stat, pvalue))

# ilk kısımda kabin numarası 1 olanların survivedlarının sumını aldım yani kabin numarası olup hayatta olan kaç kişi var, 
#sonra kabin numarası 0 olanların da sumını aldım yani kabin numarası olmayıp hayatta kalan kaç kişi var. 
# oran testi yapmak için gerekli 2 parametre var birincisi başarı sayısı ikincisi ise gözlem sayısı
# dolayısıyla bu problem için başarının tamı kabin numarası olanlar için hayatta kalmaktır, kabin numarası olmayanalrın da hayatta kalan kişilerinin toplamını bulduk
# ve nobs yani gözlemlenme frekansları var. yani kabin numarası olanlar kaç kişi, kabin nnumarası olmayanlar kaç kişi
# bunları da nobs argümanına girdik.
```

dikkat oldukça güzel bir noktadayız. proportions z testinin sonucu şu şekildedir: p1 ve p2 oranları arasında fark yoktur der. p1 ve p2 oranları neyi ifade eder. iki grubun yani kabin numarası olan ve olmayanların hayatta kalıp kalmama durumunu ifade ediyor bu oranlar. ikisi arasında fark yoktur diyen h0 hipotezi p-value değeri 0.05 ten küçük olduğundan dolayı reddedilir. yani aralarında istatistiki olarak anlamlı bir farklılık var gibi gözüküyor.

bir değişken türettik ve bu değişken benim için anlamlı olabilir mi olmayabilir mi bilmiyordum. modelleme yapmadan bunu tam anlamıyla göremem ama bana bir fikir vermesi açısından bu oranlar frekanslar da bunun dağılımı da göz önünde bulundurulduğunda acaba birbirinden farklı mıdırı sordum ve farklıdır yanıtını aldım. iki kere dikkat burada çok değişkenli etkiyi bilmiyorum. burada iki tane değişkenin sanki ikisi birlikte oluşmuş gibi değerlendirerek getirdim. ama bu yapılar tek başına oluşmadı ki yani survived değişkeninin ortaya çıkışı burada ele aldığımız new kabin bool değişkenile ortaya çıkmadı ki am çıkmış da olabilir bilmiyorum. ve şu andaki kanaatim de çıkmış olabileceğine yönelik ama bilmiyorum. dolayısıyla neyi ifade etmeye çalışıyorum. çok değişkenli etkiyi bilmiyorum ben bu değişkenlerle modele soktuğumda bu çok değişkenli etkiyi gözlemleme imkanım olacak. o durumda bu değişkenin anlamlı olup olmadığını daha iyi değerlendiriyor olacağım(new kabin bool değişkenini). ama şu anda ilerlemek ya da bu değişkeni kabul edip etmemek için bir fikir üretmeye çalıştığımda yeterli ispatı yeterli delili buldum. ancak tekrar dikkat çok değişkenli etkiyi daha bilmediğim için bu benim için önemli bir değişkendir genellemesini henüz yapamıyorum, gözönünde bulunduruyorum sadece en azından kayda değer diyorum.

```python
# şimdi binary feature oluşturmaya devam ediyorum

df.loc[((df["SibSp"] + df["Parch"]) > 0), "NEW_IS_ALONE"] = "NO" # bu veri setinde sibsp  ve parch olmak üzere iki değişken var bunlar yakın ve uzak olabilecek akrabalıkları ifade ediyor.yani kişinin gemide akrabalıklarını ifade eden iki değişken var.bunların toplamı diyorum eğer 0dan büyükse yeni bir değişken oluştur ve demek ki bu kişi yalnız değilmiş diyerek new_is_alone değişkenine no şeklinde yeni bir değer ata diyorum
df.loc[((df["SibSp"] + df["Parch"]) == 0), "NEW_IS_ALONE"] = "YES" # eğer bunların toplamı da 0a eşitse yes ata. burada olası featureları türetmeye çalışıyorum hala.
#yorumum şu belki kişi gemide tek olup olmamasına göre hayatta kalma refleksleri şekillenmişştir. gemide annesi babsı ya da çocuklarının olması onun hayatta kalma reflekslerini etkilemiş olabilir.
# yaptığım işlemde bir sonuç çıkmayabilir de ama denemem lazım. değişken üretmenin güç noktası da bu olası tüm değişkenleri üretmeye çalışıyprum.
#yaptığım işlemi survived açısından da değerlendirmem lazım

df.groupby("NEW_IS_ALONE").agg({"Survived": "mean"})
```

evet aralarında bir fark var gibi gözüküyor. yalnız mı hayır yanında ailesi, akrabası var değişkenine baktığımızda hayatta kalma oranı daha fazla gibi duruyor. bu etken şans eseri ortaya çıkmış olabilir çok değişkenli başka etkiler de olabilir sadece ama ucundan fikir edinmek için baktığımda fark var gibi gözüküyor. çünkü yalnız olanların hayatta kalma oranı daha düşük. buraya da bir hipotez testi yapacağım.

```python
test_stat, pvalue = proportions_ztest(count = [df.loc[df["NEW_IS_ALONE"] == "YES", "Survived"].sum(),
                                              df.loc[df["NEW_IS_ALONE"] == "NO", "Survived"].sum()],
                                     nobs = [df.loc[df["NEW_IS_ALONE"] == "YES", "Survived"].shape[0],
                                            df.loc[df["NEW_IS_ALONE"] == "NO", "Survived"].shape[0]])

print("Test Stat = %.4f, p-value = %.4f" % (test_stat, pvalue))

# yukarıda yaptığım işlemin aynısı 
# fark var gibi gözüküyor demiştik ve istatistiki olarak anlammlı bir fark mı bu diye bakıyorum
```

evet farklı gibi gözüküyor, h0 hipotezi reddedilir. h0 hipotezi der ki iki oran arasında fark yoktur ama reddildi varmış. bu neyi ifade eder, ilerlemek için iyi bir noktadayım, mutlak olarak hayatta kalmak için bu değişkenin etkili olduğunu göstermez ama bu değişkenin gözardı edilemeyeceğini gösterir, peki bunun etkisini daha etkili nasıl görürüz? modelleme bölümünde göreceğiz. bir model kurulduğunda oradaki etki daha net gözüküyor olacaktır.

### 3- Text Featurs

bu bölümde metinler üzerinden özellikler türetmeye çalışacağız.

```python
df.head() #en sonunda gördüğüm bütün teknikleri veri önişleme feature engineering hepsini veri setine baştan uyguluyor olacağım.
#verimi ilk haliyel kullanmak istediğim için bu sefer kernel atmadım sadece veriyi okuttum
```

```python
#buradaki metinlerden işlemler türetmek istiyorum. normalde kardinalitesi yüksek
#çöp olan bir veri olabilir name kısmı ancak bi kontrol ediyorum. çeşitli değişkenler türetmeye çalışacağım
```

```python
# 1- letter count (harflari saydırma)
# örneğin buradaki harfleri saydırarak burada kaç tane harf olduğuna bakabiliriz
#burada yorum yapıyorum ve incelemeye çalışıyorum, neler çıkabileceğini görmeye çalışıyorum.
#örneğin bazıları kraliyet ailesi soyundandır dıdısının dıdısının dıdısının dıdısı olabilir.
#bu yüzden belki de kraliyet ailelerine yapılan bir ayrıcalık vardır ve bu buraya yansımış olabilir.
#normal hayatta o kadar ilginç değişkenler türüyor ki bu şekilde probleme yaratıcı olarak yaklaşmakta fayda vardır.


df["NEW_NAME_COUNT"] = df["Name"].str.len()  #string diyorum ve len metodunu uyguluyorum.
```

```python
df.head() #bütün isimler sayıldı ve gözlemlendi ve değişken oluştu kaç harf varsa
#gözlemlenmiş oldu. harfleri saydım kelimeleri de sayabilirim
```

```python
# word count (kelime sayıları)
# apply ile lamba tanımlaması yapacağım

df["NEW_NAME_WORD_COUNT"] = df["Name"].apply(lambda x: len(str(x).split(" ")))
#bu lambda şunu yapsın ilgili ismi yakaladığında bu ismi komple stringe çevirsin daha sonra bunu split et boşluklarından ve kaç tane kelime varsa say
```

```python
df.head() #kelimeleri saydırdık sonuçlar tabloya eklendi. meta bilgileri topladım artık daha detaya inebilirim
```

```python
# özel yapıları yakalamak

# örneğin isimlerin içerisindeki doktor ifadeleri vardı örneğin bu ifadeleri flagleyebilirim.
#aslında burada mesleklerle ilgili bir değişken yok ama bene tüm featureları türetmeliyim. ondan sonra modelleme basamağında değerlendirebilirim.
#bu arada featureları sadeleştirme basamağım da olmalı gözden kaçmamalı, mümkün olduğunca az sayıda değişken ile yüksek başarıya erişmeye çalışıyorum.
#o yüzden label encodingte elemeler yaptım. ama Feature engineering bölümünde öyle değişkenle türeyebilir ki bunlar bir şekilde hataları düşürür yani,
#başarılı tahminler elde etme imkanını sağlayabilir. bu sebeple mümkün olduğunca kurcalıyoruz.

df["NEW_NAME_DR"] = df["Name"].apply(lambda x: len([x for x in x.split() if x.startswith("Dr")]))
# name kolonunda gez ilgili satırı split et, split ettiğinde name satırlarında bulunan her bir değişkene liste halinde erişilebilir olacak her bir hücreye, sonra gezdiğin her bir kelime dr ile başlıyorsa bunu seç sonra lene al.(bu bir yorum belki bu işlem 10 farklı şekilde de yapılabilir.)
```

```python
df.head() #headine baktığımda ilk etapta gözlemleyemedim. hızlı bir şekilde bakmak istiyorum
```

```python
#doktora göre groupbya alıp survivedın ortalamasını alıyorum gözlemlemek ve incelemek için

df.groupby("NEW_NAME_DR").agg({"Survived": ["mean", "count"]})
```

```python
# burada gözlemlendiği gibi doktor olanların hayatta kalma olasılığı daha yüksekmiş
# ancak bu kategorik değişken olduğu için frekanslarını gözlemlemeye çalışacağım, bu yüzden meanin yanına countu ekliyorum
# 10 tane dr varmış. 881 gözlemin olduğu yerde 10 tane dr olması ne kadar önemli yoruma açık olmakla birlikte en azından kayda değer bir sonuç elde ettik. 
```

bu bölümde regular exprection ile değişken türetme işlemlerini sürdüyor olacağız. isimlerin önündeki titleların önemli olduğunu biliyordum ve şimdi bunların üzerinde işlem yapıp bunları çekeceğim. 

```python
df.head()
```

```python
#titleı incelediğimizde başında ve sonunda boşluk var ilk harfi büyük sonraki harfler
#küçük ve boşluktan önce nokta var. burada bir örüntü yakalamaya çalışıyorum filtrelemek için.
#bu yüzden buna regular exprection deniyor. örneğin Mr. Mrs. Miss. başında boşluk büyük ve küçük harfler ve sonrasında nokta.
#filtremi oluştururken büyük küçük harfleri seçmesini başında boşluk ve sonunda nokta olduğunu belirtmem gerekmektedir.
```

```python
df["NEW_TITLE"] = df.Name.str.extract(" ([A-Za-z]+)\.", expand = False) #extract çıkar diyorum. peki nasıl çıkarmak istersin önünde boşluk olacak sonunda nokta olacak ve büyük ya da küçük harflerden oluşacak şekilde göreceğin bu ifadeleri yakala
```

```python
df.head()
```

```python
df[["NEW_TITLE", "Survived", "Age"]].groupby(["NEW_TITLE"]).agg({"Survived": "mean", "Age": ["count","mean"]})
#tıtleı seç survivedı ve agei seç, tıtlea göre grupbya al sonra survivedın ortlamasını agein toplam sayısını ve ortalamsını al
```

öyle değerlerli bir bilgi elde ettik ki gözlenme ihtimali olmayan bir değişken yani bizim için dikkate alınma ihtimali olmayan bir değişkenden dahi çok değerli bir bilgi çıkardık. burada örnğin eskiden boş değerlere yaş değerlerinin ortalamasını atamıştık ama artık bunun böyle yapılmaması gerektiğini biliyorum. örneğin burada master sınıfının ortalama yaş değeri 4müş, miss sınıfında 146 kişi var ve yaş ortalaması 21miş, mr sınıfının yaş ortalaması 32ymiş. şu anda çok değerli bir bilgiye sahibim. veri setimde daha önce olmayan bir değişken türettim bu değişken üzerinden yaşarın ortalamasını aldığımda gördüm ki hani kategorik değişken kırılımında eksiklikleri atayalım demiştik ya buna hizmet edebilecek bir değişken türedi. şimdi bu kategorik değişken kırılımında ortalamaları doldursam daha mantıklı olacak. bu şekilde olası ihtimallerde değşiken türetmenin ne kadar faydalı olabileceği gözlemlenmiştir. yine burada agein frekansını aldık bu bize bir bilgi sunuyor ve survivedın ortalamasını aldık hayatta kalma kalmama durumunu ortaya koyması için bunu da gözlemlemek için yaptık.

```python
dff = pd.read_csv("course_reviews.csv")

dff.head()
#kursun puanlamaları var, puanlamaların tarihleri, kursa puan veren kiişnin kursa iye olma tarihi, ilerleme durumu, sorduğu sorular, yanıtlanan sorular gibi değişkenleri var
# amacım timesstamp değişkeninden yeni değişkenler türetmek.
```

```python
dff.info() #ancak bir problemim var üzerinden yeni değişkenler üreteceğim değişkenin türü datetime olması gerekirken object gözüküyor.
# bu değişkenin tipini düzletmem gerekmektedir.
```

```python
dff["Timestamp"] = pd.to_datetime(dff["Timestamp"], format = "%Y-%m-%d %H:%M:%S") #to_datetim metodunu kullanarak bunu gerçekleştirebiliriz,
#bana dönüştürmek istediğin object verini ver der ve bu değişkenin içinde çeşitli tarih zaman ifadeleri var ya bunların da ne sırada
#olduğunnu bana ver der, ofrmat kısmında da bu yazmaktadır year-month-day şeklinde 
```

```python
dff.info() #aldığım hatada dattime modülüne dönüştürmek için veride saat için ayar yapmamı istedi hmsyi yazmadan hata verdi
```

```python
#yıl değişkeni türetmek istiyorum

dff["year"] = dff["Timestamp"].dt.year # dt burada datetime modülünü kastetmektedir. dt.year diyerek yılı çek diyorum.
```

```python
# ay değişkeni oluşturuyorum.
dff["year"] = dff["Timestamp"].dt.month
```

```python
#bugünün tarihinden çıkarabilirim
#veri setindeki değişkenleri bugünün yılından çıkar diyebiliyorum ve yılların arkı geliyor
dff["year_diff"] = date.today().year - dff["Timestamp"].dt.year
```

```python
#iki tarih arasındaki farkın ay cinsinden ifade edilmesi

dff["month_diff"] = (date.today().year - dff["Timestamp"].dt.year) * 12 + date.today().month - dff["Timestamp"].dt.month
#iki tarih arasındaki ay farkını hesaplamak istediğimizde öncelikle yıl farkını hesaplamalıyız sonra
#kaç yıl fark varsa onu 12 ile çarparak ay cinsine çevirmeliyiz. daha sonra iki trih arasındaki ay farkını almalıyız.
#date.today ile şu anın tarihi getiriliş sonra datadaki yıl alınmış ve bunların farkı alınmış ve 12 ile çarpılmış ve iki ay farkı ile toplanmış
# 2 * 12 = 24 ve geri kalan işlemler
```

```python
#veri setindeki günlerin hangi gün olduğuna erişmek istersek

dff["day_name"] = dff["Timestamp"].dt.day_name()
```

### 6- Özellik Etkileşimleri (Feature Interaction)

özellik etkileşimi ne demektir? değişkenler birbiri ile etkileşime girmesi demektir. yani örneğin iki değişkenin çarpılmasıdır. iki değişkenin toplanmasıdır. bir değişkenin karesinin alınması küpünün alınması gibi değişkenlerle etkileşimde olmak demektir.

```python
#örneğin yaş değişkeni ile pclassı çarpabiliriz.

df["NEW_AGE_PCLASS"] = df["Age"] * df["Pclass"] #bu interactionların teorisinde bir şey olmalı, bir şey ifade ediyor olmalı, örneğin bazı startupların tek görevi feature türetmektir. 
# burada olası tüm kombinasyonları denerler çarpıp köklerini alırlar bölerler toplarlar vs gibi ve bu yaptığımız işlem (önemli nokta burası) bir şey ifade ediyor olmalı

#burada yaş değişkeni ile pclassı çarpmak neyi ifade eder. age değişkeninin aynı standart formda olması varsayımı altında örneğin pclass 1 ile 3 arasında ya yaş değşikeni da 1 ile 3 arasında olsun
# ve burada yaşı küçük olup pclasıı 1 olan bir yolcu ile yaşı büyük olup yolculuk sınıfı 3 olan bir yolcunun hayatta kalma oranında fark olacaktır. hangi açıdan refah açısından, ister aşlesnden kaynakklı o gemiye binebiliyor olsun ister 
#kendi imkanlarıyla binebiliyor olsun dolayısıyla yaş değişkeni ile pclass değişkeninin çarpımı neticesinde gelen değerin küçük olması bir şey ifade ediyor olacaktır, büyüj olması da bir şey ifade ediyor olacaktır. mesela bu interaction, bu etkileşim teorik olarak bir şeye karşılık geliyor.
#belki de bu değişken modelleme esasında anlamlı bir yapı sahibi olacak. o yüzden üretmeye çalışıyorum. arkasındaki mantık yolcuların yaşlarının yolculuk sınıflarına göre refah durumlarıyla ilgili bir durum ortaya çıkarmak. işete genç yaşta 1. sınıfta yolculuk yapıyor ya da 3. sınıfta yolculuk yapıyor acaba refah durmumu düşük gibi bir bilgi taşıyor olabilir. 
# tekrar altınız çizmekte dayda var yaş değişkeninin bir standartlaştırılmaya tabi tutulması gerekebilir.

df["NEW_FAMILY_SIZE"] = df["SibSp"] + df["Parch"] + 1
# ya da akrabalık ilişkilerini toplayıp bir de kişiinin kendisini toplayınca family size adlı bir değişkene ulaşılabilir.


df.loc[(df["Sex"] == "male") & (df["Age"] <= 21), "NEW_SEX_CAT"] = "youngmale"  # cinsiyeti erkek olanları ve yaşı 21den küçük olanları yeni bir değişkene at ve bunları böyle isimlendir
#belirli kategorik değişken ya da sayısal değişkenlerin etkileşimleri yani etkileşim noktalrına da flagler atılabilir.burada değişkenler oluşturulabilir.

#aynı değişkene başka bir sınıf oluşturalım
df.loc[(df["Sex"] == "male") & ((df["Age"] > 21) & (df["Age"]) <= 50), "NEW_SEX_CAT"] = "maturemale" # new_sex_cat katgorisine  erkek olup yaşı 21den büyük ve 50den küçük olanları maturemale şeklinde isimlendir diyorum yani olgun erkekler olarak.

#daha yaşlı olanlara ise:
df.loc[(df["Sex"] == "male") & (df["Age"] > 50), "NEW_SEX_CAT"] = "seniormale" #şeklinde isimlendirebilirim.
# hala aynı değişkendeyim yakaladığım bazı yerlere göre isimlendirmeler yapıyorum.

#kadınlar için:
df.loc[(df["Sex"] == "female") & (df["Age"] <= 21), "NEW_SEX_CAT"] = "youngfemale"

#yukarıda yaptığım işlemlerin aynısını kadınlar için yapıyorum. olgun kadınlar:
df.loc[(df["Sex"] == "female") & ((df["Age"] > 21) & (df["Age"]) <= 50), "NEW_SEX_CAT"] = "maturefemale"

#daha olgun kadınlar

df.loc[(df["Sex"] == "female") & (df["Age"] > 50), "NEW_SEX_CAT"] = "seniorfemale"
```
