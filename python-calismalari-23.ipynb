```python
import matplotlib.pyplot as plt
import  numpy as np
```

```python
x = [1,2,3,4]
y= [1,4,9,16]

plt.plot(x,y)
plt.show() #verdiƒüimiz x deƒüerleri x koordiatƒ± √ºst√ºnde olu≈üturuldu 
```

```python
x = [1,2,3,4]
y= [1,4,9,16]

plt.plot(x,y)
plt.axis([0,6,0,20]) #hangi aralƒ±klarda olmasƒ± gerektiƒüini s√∂ylebilirim deƒüerlerin ilk verdiƒüim x son verdiƒüim y deƒüerlerinin aralƒ±ƒüƒ±nƒ± ifade ediyor
plt.show()
```

```python
x = [1,2,3,4]
y= [1,4,9,16]

plt.plot(x,y, color = "red", linewidth = 5) #renk ekleyebilirim color ile ve linewidth ile √ßizgi kalƒ±nlƒ±ƒüƒ±nƒ± deƒüi≈ütirebilirim 
plt.axis([0,6,0,20])

plt.title("Grafik ba≈ülƒ±ƒüƒ±") #ba≈ülƒ±k ekleyebilirim
plt.xlabel("x label") #x eksenine isim veririm
plt.ylabel("y label") #y eksenine isim veririm
plt.show

# matplotlib style yazarsam googlea stiler hakkƒ±nda bilgi bulabilirim
```

```python
x = [1,2,3,4]
y= [1,4,9,16]

plt.plot(x,y,"o--r") #direkt renk i√ßin harfleri de kullanabilirim ve -- dediƒüimde kesikli olarak bana verir, o dediƒüimde de yuvarlak koyar, ilk dediƒüim marker ikinci √ßizginin tipi √º√ß√ºnc√º renk
plt.axis([0,6,0,20])

plt.title("Grafik ba≈ülƒ±ƒüƒ±") #ba≈ülƒ±k ekleyebilirim
plt.xlabel("x label") #x eksenine isim veririm
plt.ylabel("y label") #y eksenine isim veririm
plt.show
```

```python
x = np.linspace(0,2,100)

plt.plot(x, x, label = "lineer", color = "red")
plt.plot(x, x**2, label = "quadratic", color= "yellow")
plt.plot(x, x**3, label = "cubic", color = "green")

plt.xlabel("x label")
plt.ylabel("y label")

plt.legend() #labellerda vermi≈ü olduƒüum deƒüer g√∂sterilmi≈ü olacak yani yukarƒ±da √ßƒ±kan k√º√ß√ºk kutucuk

plt.title("simple plot")

plt.show()

# d√ºzleme birden fazla grafik ekleyebiliyorum
```

```python
x = np.linspace(0,2,100)

fig, axs = plt.subplots(3) #subplots metoduna 2 parametresini verdiƒüimde 2 axes alanƒ± olu≈üturmu≈ü oluyor ve iki farklƒ± grafik g√∂rm√º≈ü oluyorum
axs[0].plot(x, x, color = "red") #axs[0] dediƒüimde ilk grafik 1 dediƒüimde ikinci grafik oluyor
axs[0].set_title("linear") #set_title metodu ile grafiklerime isim ekleyebilirim

axs[1].plot(x, x**2, color = "green")
axs[1].set_title("quadretic")

axs[2].plot(x, x**3, color = "yellow")
axs[2].set_title("cubic")

plt.tight_layout() # dediƒüimde de grafikler birbirinden uzakla≈üƒ±r

plt.show()
```

```python
#alt alta yerine d√∂rt b√∂lmeli grafik yapabilirim 

x = np.linspace(0,2,100)

fig, axs = plt.subplots(2,2) # dediƒüim zaman 2 ye 2 toplamda 4 grafik olu≈üur 
fig.suptitle("grafiƒüin ana ba≈ülƒ±ƒüƒ±") # t√ºm√ºn√ºn ana ba≈ülƒ±ƒüƒ± i√ßin fig subtitle diyorum

axs[0,0].plot(x, x, color = "red") #0a 0 lƒ±k alan
axs[0,1].plot(x, x**2, color = "blue") #0a 1lik alan
axs[1,0].plot(x, x**3, color = "green") #1e 0lƒ±k alan
axs[1,1].plot(x, x**4, color = "yellow") #1e 1lik alan
```

#  21.3 Matplotlib ile figure olu≈üturma

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =np.linspace(-10,9,20) #aralƒ±kta olu≈üturulan sayƒ±lar -10 ile 9 arasƒ±nda 20 sayƒ±
y = x **3 #sayƒ±larƒ±n k√ºb√ºn√º alƒ±yor
z = x **2

figure = plt.figure() #figure olu≈üturdum

axes_cube = figure.add_axes([0.1, 0.1,0.9,0.9]) #add_axes metodu figure eklenecek olan axes alanƒ±nƒ±n konumunu belirtiyor soldan 0.2 olsn a≈üaƒüƒ±dan 0.2 olsun geni≈ülik 0.8 olsun ve y√ºkseklik de 0.8 olsun dedim visualda yaptƒ±ƒüƒ± i√ßin o grafiƒüin ekran √ºzerindeki konumuyla oynuyor ≈üu an bende deƒüi≈ümedi
axes_cube.plot(x,y,"b")
axes_cube.set_xlabel("x axis")
axes_cube.set_ylabel("y axis")
axes_cube.set_title("cube")


axes_square = figure.add_axes([0.2,0.7,0.3,0.3]) #yeni eklediƒüim akses ve konumu grafiƒüin i√ßine doƒüru ayarladƒ±m
axes_square.plot(x,z,"r")
axes_square.set_xlabel("x axis")
axes_square.set_ylabel("y axis")
axes_square.set_title("square")

plt.show()
```

```python
figure = plt.figure()

axes = figure.add_axes([0,0,1,1])

axes.plot(x,z,label = "square")
axes.plot(x,y,label = "cube")
axes.legend(loc = 5) #aynƒ± akses i√ßinde grafiƒüimizi √ßizdik bunun ile bilgilendirme sol k√∂≈üede belirdi loca sayƒ± verince yeri deƒüi≈üiyor

plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols =1,figsize=(4,4)) #bu prametreleri vermek zorunda deƒüilim yine aynƒ± ≈üeyi ifade ediyor, figsize grafiƒüin boynu ayarlƒ±yor
axes[0].plot(x,y)
axes[0].set_title("square")

axes[1].plot(x,z)
axes[1].set_title("cube")

plt.tight_layout()
#fig.savefig("figure1.png")# figuru kaydetmek i√ßin kullanƒ±lƒ±r png yerine pdf de yazabilrim

plt.show()
```

# 21.4 matplotlib ile grafik t√ºrleri

```python
import matplotlib.pyplot as plt
```

```python
yil = [2011,2012,2013,2014,2015]

oyuncu1 = [8,10,12,7,9]
oyuncu2 = [7,12,5,15,21]
oyuncu3 = [18,20,22,25,19]

# stack plot yƒ±ƒüƒ±n grafik grafik kullanacaƒüƒ±z

plt.plot([],[],color = "orange",label = "oyuncu1")#ba≈üta bo≈ü listeler veriyorum daha sonra a≈üaƒüƒ±da bilgilerimi veriyorum
plt.plot([],[],color = "r",label = "oyuncu2")
plt.plot([],[],color = "blue",label = "oyuncu3")

plt.stackplot(yil, oyuncu1,oyuncu2,oyuncu3, colors=["y","r","b"])
plt.title("yƒ±llara g√∂re ayƒ±lan goller")
plt.xlabel("yil")
plt.ylabel("gol sayƒ±sƒ±")

plt.legend()
plt.show()
```

```python
# pasta grafik

goaltypes = "penaltƒ±","kaleye ≈üut", "serbest vuru≈ü" #gol t√ºrlerimi yazdƒ±m

goals = [12,35,7] #gol sayƒ±larƒ±mƒ± yazdƒ±m
colors = ["y","r","b"] #her bir pasta dilimi i√ßin renk belirttim

plt.pie(goals, labels=goaltypes,colors=colors, shadow=True,explode=(0.05,0.05,0.05), autopct="%1.1f%%") #pie demek pasta grafiƒüi √ßaƒüƒ±rƒ±yor labela goaltypes oturacak ve renklere de tanƒ±mladƒ±ƒüƒ±m renkler parametresini veriyorum
#shadow g√∂lge olsun mu, explode her bir dilim arasƒ±nda mesafe olsun mu, ve autopct y√ºzdelik dilimleri g√∂steriyor ve y√ºzdelik ifadenin formatƒ±nƒ± yazƒ±yorum kar≈üƒ±sƒ±na 

plt.show()
```

```python
#bar grafik

plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20], label = "BMW", width=.5) #birincisi g√ºn bilgisi ikincii mesafe bilgisi olacak
plt.bar([0.75,1.75,2.75,3.75,4.75],[80,20,20,50,60], label = "audi", width=.5) #width grafiklerin kalƒ±nlƒ±ƒüƒ±nƒ± ayarlƒ±yor

plt.legend()

plt.xlabel("g√ºn")
plt.ylabel("mesafe km")

plt.title("ara√ß mesafe bilgileri")
plt.show()
```
